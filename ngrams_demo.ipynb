```python
"""
Input is an article which data type is String.
Output is a list of binary-tuple which contains “the term generated by the method” and “the frequency of the corresponding term”.
Set N from 2 to 5 for N-grams.
Only output the binary-tuple where its “the term generated by the method” is the “longest common substring”.
Only output the binary-tuple where its “frequency of the corresponding term” is more than or equal to 3. 
"""
```

```python
# load text

text = open('./trump_appledaily_20170117.txt', 'r').read()
```

```python
# keep zh-tw char

import re
clean_text = ("".join(re.findall(u"[\u4e00-\u9fa5]+", text)))
```

```python
clean_text
```

```python
# ngrams func

def ngrams(input_text, n): 
    for i in range(len(input_text)-n+1):
        yield input_text[i:i+n]
```

```python
print(" ".join(ngrams(clean_text,2)))
```

```python
from collections import Counter

uni = Counter(ngrams(clean_text, 1))
bi = Counter(ngrams(clean_text, 2))
tri = Counter(ngrams(clean_text, 3))
quad = Counter(ngrams(clean_text, 4))
qui = Counter(ngrams(clean_text, 5))
```

```python
uni.most_common(50)
```

```python
bi.most_common(50)
```

```python
tri.most_common(50)
```

```python
quad.most_common(50)
```

```python
qui.most_common(50)
```

```python
# setup stop word

import numpy

stopword_threshold = numpy.mean(list(uni.values())) + 2 * numpy.std(list(uni.values()))
stopword_threshold
```
